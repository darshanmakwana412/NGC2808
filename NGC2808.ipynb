```python
# Importing Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
%matplotlib inline
```

```python
# Loading the txt Data
txtData = np.loadtxt('NGC2808.txt', dtype='str')

# Creating a pandas Dataframe from the loaded txt data
data = pd.DataFrame(txtData)
data[data.loc[:, data.columns != 35].columns] = data[data.loc[:, data.columns != 35].columns].apply(pd.to_numeric)
```

```python
#Getting info of dtypes for the data
data.info()

#printing first 5 rows of the data
data.head()
```

```python
# Plotting a scatter plot of F275W versus F275W-F336W
plt.ylim(30, 15)
plt.xlim(-5, 5)
sns.set(rc = {'figure.figsize':(15,15)})
sns.scatterplot(data[2]-data[8], data[2], alpha=0.03, marker='.')
plt.show()
```
